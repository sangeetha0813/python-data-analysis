## Level 1: Data Inspection 

```python
import numpy as np
import pandas as pd

# Make output reproducible
np.random.seed(42)

# Number of records
n = 100

# Generate random data
df = pd.DataFrame({
    "Order_ID": np.arange(1001, 1001+n),
    "Customer": np.random.choice(
        ["Amit", "Priya", "Rahul", "Neha", "Rohan", "Anjali", "Vikram", "Sneha", "Karan", "Meera"],
        n
    ),
    "City": np.random.choice(
        ["Delhi", "Mumbai", "Bangalore", "Hyderabad", "Pune", "Chennai"],
        n
    ),
    "Category": np.random.choice(
        ["Electronics", "Furniture", "Clothing"],
        n
    ),
    "Product": np.random.choice(
        ["Laptop", "Chair", "Mouse", "Table", "Keyboard", "Monitor",
         "Shoes", "Jacket", "Mobile", "Desk"],
        n
    ),
    "Quantity": np.random.randint(1, 6, n),
    "Price": np.random.randint(500, 50001, n),
    "Payment_Mode": np.random.choice(
        ["UPI", "Card", "COD", "EMI"],
        n
    ),
    "Status": np.random.choice(
        ["Delivered", "Pending", "Cancelled"],
        n,
        p=[0.7, 0.2, 0.1]
    ),
    "Discount": np.random.randint(0, 3001, n)
})

# Create Sales column
df["Sales"] = df["Quantity"] * df["Price"] - df["Discount"]

# Add some missing values
df.loc[np.random.choice(df.index, 5, replace=False), "Discount"] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), "City"] = np.nan

# Rearrange columns
df = df[
    [
        "Order_ID", "Customer", "City", "Category", "Product",
        "Quantity", "Price", "Discount", "Sales",
        "Payment_Mode", "Status"
    ]
]

# Save as CSV
df.to_csv("sales_data.csv", index=False)
```
